```python
import tensorflow as tf
from keras.models import load_model
from collections import deque
import numpy as np
import pickle
import cv2 as cv
```

```python
model_path = './model/theftSurvModel' #path to your theftsurvmodel directory
model = tf.keras.models.load_model(model_path)
#model = load_model('./model/theftSurvModel/')
path_to_output = ('./output')
lb = pickle.loads(open('./model/theftSurvLB.pickle', 'rb').read())
mean = np.array([123.68, 116.779, 103.939][::1], dtype = 'float32')
Queue = deque(maxlen = 128)
```

```python
capture_video = cv.VideoCapture(0)
writer = None
(Width, Height) = (None, None)

while True:
    (taken, frame) = capture_video.read()
    if not taken:
        break
    
    if Width is None or Height is None:
        (Width, Height) = frame.shape[:2]
        
    output = frame.copy()
    frame = cv.cvtColor(frame, cv.COLOR_BGR2RGB)
    frame = cv.resize(frame, (64,64)).astype('float32')
    frame -= mean
    preds = model.predict(np.expand_dims(frame, axis = 0))[0]
    Queue.append(preds)
    results = np.array(Queue).mean(axis = 0)
    i = np.argmax(results)
    label = lb[i]
    text = 'Theft classified : {}'.format(label)
    cv.putText(output, text, (45, 60), cv.FONT_HERSHEY_PLAIN, 1.25, (255,0,0), 2)
    
    if writer is None:
        fourcc = cv.VideoWriter_fourcc(*'MJPG')
        writer = cv.VideoWriter(path_to_output, fourcc, 30, (Width, Height), True)
    writer.write(output)
    cv.imshow('In progress', output)
    key = cv.waitKey(1) & 0xFF
    
    if key == ord('q'):
        break
    
print('Saving......')
writer.release()
capture_video.release()
cv.destroyAllWindows()
```
